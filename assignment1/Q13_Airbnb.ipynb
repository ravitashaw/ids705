## Neighbourhood

There are more than 50000 Airbnb listings in New York City. A majority of these listings are located in Manhattan and Brookyln. The neighbourhood of Queens, Bronx, Staten Island have relatively lower number of listings. Becasue of the soaring prices of hotels in the city, Airbnb has made the life of short term travellers easier. A hotel for short term will cost considerably a lot and can be tiring if the stay is more than a week. With Airbnb apartments available in corporate housing area, it is easy to book and traveller can know what to expect. Hotels are predictable and non Airbnb temporary housing solutions can be very surprising as there is no trace of reviews available online. In addition, those accomodations are hard to reserve and may lead to even cheating or scams. Airbnb acts as a medium for the owner and travellers and has reviews for different categories which provides a relatively transparent picture of what to expect. Because of this transparency, the listing owners try to provide their best. If the ratings of the listing will go down, less people would like to make reservation for those listings. Considering the high rental cost of apartments in New York, the listing owner rely heavily on earning money from subletting a part of the apartment. So, this model works for both the owners and the travellers. 

Manhattan has the highest number of listings where Entire Home or apartment is available for the traveller. Brooklyn, and all other neighbourhoods have eithe almost equal number of entire home apartment and private room listings. This trend could be a result of the fact that there are more studio or 1 bed room apartments in Manhattan and New Yorkers travelling elsewhere would rent out their entire house on Airbnb. On the contrary, Queens and Bronx have higher number of private bedroom over entire house apartments. These areas are relatively less popular than Manhattan and Brooklyn and the apartments in these areas are usually bigger. So, people would prefer to rent out these one of the several bedrooms they might have. There are very small number of listings of type Hotel in all the neighborhood and does not seem to be very popular among Airbnb customers. The most significant reason for using Airbnb is to not have a hotel lke experience and therefore, people do not prefer to book places which are of type hotel. 

## Availability

A majority of listings that are available for less than 1 month are Entire Home or Apartments. This indicates that it could be due to that a New Yorkers list their place on accomodation for short duration when they are travelling. However, there is not sufficient data to claim that the listings are available in continous span of time. It could be possibly true that owners travel for a week several times during an year and are available for booking only during those discrete number of weeks. So, making an assumption based upon the dataset might be limiting the analysis for availability. This reasoning is more plausible for listings which are available for less than a month. 

The private room are available in the same proportion throughout the year. Considering the high cost of real estate in New York and the hastle of finding a roomate or signing a lease with the owner. Airbnb listings save people from all these hastles and provides respectful and easy going travellers and owners. Also, it is temporary accomodation so a person is not bound to stay for the contract like a typical housing lease aggrement. 

There are very few proportion of listings where a traveller will have to share a room with someone. People dont like to be intruded by privacy usually and so a lot of neither traveller nor owners would prefer a shared room with a stranger. However, there are both travellers and owners from all segments of society and both of these groups of people might have financial challenges and so they would want to try to save as much as possible, which can be very hard in a city like New York, by best utlization of resoruces which could result into the sharing room listings.

## Ratings

There are several type of rating scores available in the dataset, including the overall rating score. It can be observed that the scores that matters the most for overall rating is the rating for communication, followed by ratings for cleanliness. Both of these factors are certainly one of the most important criteria for booking a listing on Airbnb. 

The average reviews are good for a lot of Airbnb listings and has a varying pattern depending upon the room type. The Entire Home/Apartment and private room type listings usually have good reviews in Manhattan and are not fundamentally very different. However, for all other neighbourhoods, the private room listings seems to have more variation in the reviews. This could be an indicator that the living in a private room with shared space like kitchen and living room are comfortable and compromising who live in Manhattan and not so compromising in other neighbourhoods.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")
sns.set_context({"figure.figsize": (8, 5)})

reviews = pd.read_csv('./data/listings.csv')

reviews = reviews[['id', 'name',
       'host_id', 
       'neighbourhood', 'neighbourhood_cleansed',
       'neighbourhood_group_cleansed',
       'latitude', 'longitude',
       'property_type', 'room_type',
       'bathrooms', 'bedrooms', 'beds', 'bed_type',
       'price', 'weekly_price', 'monthly_price', 'minimum_nights',
       'maximum_nights', 'availability_365', 'number_of_reviews', 'number_of_reviews_ltm',
       'review_scores_rating', 'review_scores_accuracy',
       'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location',
       'review_scores_value','reviews_per_month']]

reviews = reviews[reviews['review_scores_rating'].isnull()==False]

reviews.loc[reviews['bathrooms'].isnull(), 'bathrooms'] = 1
reviews.loc[reviews['bedrooms'].isnull(), 'bedrooms'] = 1
reviews.loc[reviews['beds'].isnull(), 'beds'] = 1
reviews.loc[reviews['weekly_price'].isnull(), 'weekly_price'] = 0.0
reviews.loc[reviews['monthly_price'].isnull(), 'monthly_price'] = 0.0
reviews.loc[reviews['review_scores_rating'].isnull(), 'review_scores_rating'] = 0.0
reviews.loc[reviews['review_scores_accuracy'].isnull(), 'review_scores_accuracy'] = 0.0
reviews.loc[reviews['review_scores_cleanliness'].isnull(), 'review_scores_cleanliness'] = 0.0
reviews.loc[reviews['review_scores_checkin'].isnull(), 'review_scores_checkin'] = 0.0
reviews.loc[reviews['review_scores_communication'].isnull(), 'review_scores_communication'] = 0.0
reviews.loc[reviews['review_scores_location'].isnull(), 'review_scores_location'] = 0.0
reviews.loc[reviews['review_scores_value'].isnull(), 'review_scores_value'] = 0.0
reviews.loc[reviews['reviews_per_month'].isnull(), 'reviews_per_month'] = 0.0

reviews.price = reviews.price.str.replace('$', '')
reviews.weekly_price = reviews.weekly_price.str.replace('$', '')
reviews.monthly_price = reviews.monthly_price.str.replace('$', '')

reviews.price = reviews.price.str.replace(',', '').astype(float)
reviews.weekly_price = reviews.weekly_price.str.replace(',', '').astype(float)
reviews.monthly_price = reviews.monthly_price.str.replace(',', '').astype(float)

reviews.loc[reviews['weekly_price'].isnull(), 'weekly_price'] = 0.0
reviews.loc[reviews['monthly_price'].isnull(), 'monthly_price'] = 0.0

reviews['availability_tr'] = 'Less than 1 month'
reviews.loc[reviews['availability_365']<30, 'availability_tr'] = 'Less than 1 month'
reviews.loc[(reviews['availability_365']>=30) & (reviews['availability_365']<90), 'availability_tr'] = '1-3 months'
reviews.loc[(reviews['availability_365']>=90) & (reviews['availability_365']<180), 'availability_tr'] = '3-6 months'
reviews.loc[(reviews['availability_365']>=180), 'availability_tr'] = 'More than 6 months'

reviews = reviews[reviews.price< 300]
reviews = reviews[reviews.review_scores_rating>=50]
```

```python
sns.boxplot(x='neighbourhood_group_cleansed', y='review_scores_rating', data=reviews, hue='room_type')
```

```python
sns.set(style="white")
# Create a covariance matrix
corr = reviews[['review_scores_rating', 'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location']].corr()
# Creating a mask the size of our covariance matrix
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11,9))
# Generate a custom diverging colormap
#cmap = sns.diverging_palette(22,10,as_cmap=True)
cmap = sns.diverging_palette(133, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
heatmap = sns.heatmap(corr,mask=mask,cmap=cmap,vmax=1,center=0,square=True,annot=True, 
           cbar_kws={'shrink': .5})
ax.set_title('Heatmap of Important Review scores')
```
