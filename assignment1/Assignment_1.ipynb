**[3 points]** What is the cumulative distribution function (CDF) that corresponds to the following probability distribution function? Please state the value of the CDF for all possible values of $x$.

$f(x) = \begin{cases}
    \frac{1}{3} & 0 < x < 3 \\
    0           & \text{otherwise}
    \end{cases}$

**ANSWER**

CDF is the area under the probability distribution function (PDF). So, CDF can be obtained by  integrating the corresponding PDF.

Generally, if F(x) is the CDF for a corresponding PDF, f(x). The relation between PDF and CDF can be represented as:

$
\qquad F(x) = \int^{x}_{-\infty} f(x) dx
$

For this problem, if x is in the interval $[-\infty, 0]$, then

$
\qquad F(x) = \int^{x}_{-\infty} 0 dx
$

$
\qquad \qquad = 0
$

If x is in the interval $(0, 3)$, then

$
\qquad F(x) = \int^{0}_{-\infty} 0 dx + \int^{x}_{0} \tfrac{1}{3} dx  
$

$
\qquad \qquad = 0 + \tfrac{x}{3} \Big|_0^x
$

$
\qquad \qquad = \tfrac{x}{3}
$

If x is in the interval $[3, \infty]$, then

$
\qquad F(x) = \int^{0}_{-\infty} 0 dx + \int^{3}_{0} \tfrac{1}{3} dx + \int^{x}_{3} 0 dx 
$


$
\qquad \qquad = 0 + \tfrac{x}{3} \Big|_0^3 + 0
$


$
\qquad \qquad = \tfrac{3}{3} - \tfrac{0}{3}
$

$
\qquad \qquad = 1
$

Therefore, the CDF for above PDF is given by:

$
\qquad F(x) = \begin{cases}
    0 & -\infty < x \leq 0 \\
    \tfrac{x}{3} & 0 < x < 3 \\
    1 & 3 \leq x < \infty \\
    \end{cases}
$
