The function below computes the function $f(x,y) = x^2 - 2 y^2$ and determines whether this quantity is above or below a given threshold, `thresh=0`. This is done for $x,y \in \{-4,4\}$, over a 2,000-by-2,000 grid covering that domain.

(a) Vectorize this code and demonstrate (as in the last exercise) the speed increase through vectorization and (b) plot the resulting data - both the function $f(x,y)$ and the thresholded output - using [`imshow`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html?highlight=matplotlib%20pyplot%20imshow#matplotlib.pyplot.imshow) from `matplotlib`.

**ANSWER**

```python
import numpy as np
import time
import matplotlib.pyplot as plt

sns.set_style("white")
sns.set_context({"figure.figsize": (5, 5)})

nvalues = 2000
xvalues = np.linspace(-4,4,nvalues)
yvalues = np.linspace(-4,4,nvalues)
thresh  = 0

# Nonvectorized implementation
t0 = time.time()
f = np.zeros((nvalues,nvalues))
f_thresholded = np.zeros((nvalues,nvalues))
for ix, x in enumerate(xvalues):
    for iy, y in enumerate(yvalues):
        f[ix,iy]             = x**2 - 2 * y**2
        f_thresholded[ix,iy] = f[ix,iy] > thresh
        
t1 = time.time()
time_nonvectorized = (t1 - t0)/1000

# Vectorized implementation
#Start time
t2= time.time()
#Create grid for vectorizing the x and y values
xx, yy = np.meshgrid(xvalues, yvalues, sparse=True, indexing='ij')

f_thresholded_vec = xx**2 - 2*yy**2
f_thresholded_op = f_thresholded_vec > thresh 

#End time
t3= time.time()
#Calculate total time
time_vectorized = (t3 - t2)/1000

# Vectorization speed performance results
print('Time [sec] (non-vectorized): ' + str(time_nonvectorized))

print('Time [sec] (vectorized): ' + str(time_vectorized))

print('The vectorized code is ' + str(time_nonvectorized/time_vectorized) + ' times faster than the vectorized code')

# Plot the function f(x,y)
fig1 = plt.imshow(f_thresholded_vec)
plt.title("Function F(x,y) Distribution over the grid")
plt.xlabel('x')
plt.ylabel('y')
plt.show(fig1)

# Plot the function f(x,y)
fig2 = plt.imshow(f_thresholded_op)
plt.title("Threshold of F(x,y) Distribution over the grid")
plt.xlabel('x')
plt.ylabel('y')
plt.show(fig2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
# figure size in inches
rcParams['figure.dpi'] = 100

#Set style and size for all the plots and figures
sns.set_style("white")
sns.set_context({"figure.figsize": (8, 5)})

#Load the data from the csv file
reviews = pd.read_csv('./data/listings.csv')

#Select (filter) the columns to be used for study 
reviews = reviews[['id', 'name',
       'host_id', 
       'neighbourhood', 'neighbourhood_cleansed',
       'neighbourhood_group_cleansed',
       'latitude', 'longitude',
       'property_type', 'room_type',
       'bathrooms', 'bedrooms', 'beds', 'bed_type',
       'price', 'weekly_price', 'monthly_price', 'minimum_nights',
       'maximum_nights', 'availability_365', 'number_of_reviews', 'number_of_reviews_ltm',
       'review_scores_rating', 'review_scores_accuracy',
       'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location',
       'review_scores_value','reviews_per_month']]


#remove records which have null values for reviews
reviews = reviews[reviews['review_scores_rating'].isnull()==False]

#Handle Missing Data 
reviews.loc[reviews['bathrooms'].isnull(), 'bathrooms'] = 1
reviews.loc[reviews['bedrooms'].isnull(), 'bedrooms'] = 1
reviews.loc[reviews['beds'].isnull(), 'beds'] = 1
reviews.loc[reviews['weekly_price'].isnull(), 'weekly_price'] = 0.0
reviews.loc[reviews['monthly_price'].isnull(), 'monthly_price'] = 0.0
reviews.loc[reviews['review_scores_rating'].isnull(), 'review_scores_rating'] = 0.0
reviews.loc[reviews['review_scores_accuracy'].isnull(), 'review_scores_accuracy'] = 0.0
reviews.loc[reviews['review_scores_cleanliness'].isnull(), 'review_scores_cleanliness'] = 0.0
reviews.loc[reviews['review_scores_checkin'].isnull(), 'review_scores_checkin'] = 0.0
reviews.loc[reviews['review_scores_communication'].isnull(), 'review_scores_communication'] = 0.0
reviews.loc[reviews['review_scores_location'].isnull(), 'review_scores_location'] = 0.0
reviews.loc[reviews['review_scores_value'].isnull(), 'review_scores_value'] = 0.0
reviews.loc[reviews['reviews_per_month'].isnull(), 'reviews_per_month'] = 0.0

reviews.loc[reviews['weekly_price'].isnull(), 'weekly_price'] = 0.0
reviews.loc[reviews['monthly_price'].isnull(), 'monthly_price'] = 0.0

#Data Cleaning
#Remove $ and comma sign from price for comparision
reviews.price = reviews.price.str.replace('$', '')
reviews.weekly_price = reviews.weekly_price.str.replace('$', '')
reviews.monthly_price = reviews.monthly_price.str.replace('$', '')

#Convert the string type of prices to float
reviews.price = reviews.price.str.replace(',', '').astype(float)
reviews.weekly_price = reviews.weekly_price.str.replace(',', '').astype(float)
reviews.monthly_price = reviews.monthly_price.str.replace(',', '').astype(float)

#Create a new feature to categorize the availability of the listings into 4 major groups
reviews['availability_tr'] = 'Less than 1 month'
reviews.loc[reviews['availability_365']<30, 'availability_tr'] = 'Less than 1 month'
reviews.loc[(reviews['availability_365']>=30) & (reviews['availability_365']<90), 'availability_tr'] = '1-3 months'
reviews.loc[(reviews['availability_365']>=90) & (reviews['availability_365']<180), 'availability_tr'] = '3-6 months'
reviews.loc[(reviews['availability_365']>=180), 'availability_tr'] = 'More than 6 months'

#Removing Outliers
#Remove records with prices higher than $300
reviews = reviews[reviews.price< 300]
#Remove records which have reviews less than 50 out of 100
reviews = reviews[reviews.review_scores_rating>=50]
```
