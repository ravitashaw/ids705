1. The sample size $n$ is extremely large, and the number of predictors $p$ is small.
2. The number of predictors $p$ is extremely large, and the number of observations $n$ is small.
3. The relationship between the predictors and response is highly non-linear.
4. The variance of the error terms, i.e. $\sigma^2 = Var(\epsilon)$, is extremely high

**ANSWER**

1. For large sample size, a flexible learning method will do *better*. A large sample size can reduce overfitting the data and can extract more information from the larger sample size. 

2. For small number of observations, a flexible method will do *worse*. As the flexible method would do overfitting and result in high variance. 

3. For non-linear relationships, a flexible model will perform *better* to the non-linear relationship as it would result in low bias. 

4. For high variance of error terms, the flexible model would perform *worse*. A flexible model would capture too much noise in the data, therefore leading to even higher variance in the testing data.

```python
# (a) Write your own kNN classifier
from scipy import stats
import numpy as np

class Knn:
# k-Nearest Neighbor class object for classification training and testing
    def __init__(self):
        pass
        
    def fit(self, x, y):
        # Save the training data to properties of this class
        self.X_train = x
        self.y_train = y
        pass

    def predict(self, x, k):
        y_hat = [] # Variable to store the estimated class label for 
        # Calculate the distance from each vector in x to the training data
        for pt in x:
            dist = pt - self.X_train
            distances = np.sum(dist*dist, axis=1)
            #get the index of top k nearest neighbours
            #argpartition partitions the array at kth index and puts k elements smaller before and larger one after
            idx = np.argpartition(distances, k)[:k]
            nearest_y = np.take(self.y_train, idx)
            y_hat.append(stats.mode(nearest_y,axis=0)[0][0])
        y_hat = np.array(y_hat)
        
        # Return the estimated targets
        return y_hat

# Metric of overall classification accuracy
#  (a more general function, sklearn.metrics.accuracy_score, is also available)
def accuracy(y,y_hat):
    nvalues = len(y)
    accuracy = sum(y == y_hat) / nvalues
    return accuracy
```

```python
# (b) Load dataset

X_test_high = np.genfromtxt('data/A2_X_test_high.csv',delimiter=",")
X_test_low = np.genfromtxt('data/A2_X_test_low.csv',delimiter=",")
X_train_high = np.genfromtxt('data/A2_X_train_high.csv',delimiter=",")
X_train_low = np.genfromtxt('data/A2_X_train_low.csv',delimiter=",")
y_test_high = np.genfromtxt('data/A2_y_test_high.csv',delimiter=",")
y_test_low = np.genfromtxt('data/A2_y_test_low.csv',delimiter=",")
y_train_high = np.genfromtxt('data/A2_y_train_high.csv',delimiter=",")
y_train_low = np.genfromtxt('data/A2_y_train_low.csv',delimiter=",")
```

```python
# (c)
# Evaluate the performance of your kNN classifier on a low- and a high-dimensional dataset 
#   and time the predictions of each

import time

# low dimensional data
t1 = time.time()

KNN_low = Knn()
KNN_low.fit(X_train_low, y_train_low)

predictions_low = KNN_low.predict(X_test_low, 5)
accuracy_low = accuracy(y_test_low, predictions_low)

t2 = time.time()
time_low = t2 - t1
print('\nKNN Classifier (k=5) Accuracy, Low dimensional data, is {} %'.format((accuracy_low*100)))
print("Time [sec] (low-dimensional):" + str(time_low))

# high dimensional data
t3 = time.time()

KNN_high = Knn()
KNN_high.fit(X_train_high, y_train_high)

predictions_high = KNN_high.predict(X_test_high, 5)
accuracy_high = accuracy(y_test_high, predictions_high)

t4 = time.time()
time_high = t4 - t3
print('\nKNN Classifier (k=5) Accuracy, High dimensional data, is {} %'.format((accuracy_high*100)))
print("Time [sec] (high-dimensional):" + str(time_high))
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from sklearn.neighbors import KNeighborsClassifier
import random
from matplotlib.colors import ListedColormap

import matplotlib.pyplot as plt

# a
X_train,y_train = make_moons(n_samples=1000, shuffle=True, noise= 0.35, random_state=1)

# b
fig = plt.figure(figsize=(10,10))
scatter_plot = plt.scatter(X_train[:,0], X_train[:,1], marker='x', c=y_train)

legend = plt.legend(*scatter_plot.legend_elements(num=1))
fig.add_artist(legend)


fig.suptitle('Scatterplot of Synthetic Dataset')
fig.text(0.5, 0.5, s='X1')
fig.text(0.5, 0.5, s='X2')
plt.show()
```

```python
# d
import matplotlib.pyplot as plt
h = .01 # step size of the mesh

fig, axs = plt.subplots(3,3, figsize=(15, 15), facecolor='w', edgecolor='k')

# X values in list from randomly generated data
X = [X_1, X_25, X_50]
# Y values in list from randomly generated data
Y = [Y_1, Y_25, Y_50]
# KNN models with different k values in list 
KNN = [knn_1, knn_25, knn_50]
# Plot number on the grid
plt_idx = 1

#iterate through all the X and Y pairs
for idx in range(len(X)):  
    # For each X and Y pair, iterate through all the KNN models
    for idx_k in range(len(KNN)):
        #specify location of the plot on grid
        ax = plt.subplot(3, 3, plt_idx) 

        Xi = X[idx]
        Yi = Y[idx]
        knni = KNN[idx_k]

        # https://www.kaggle.com/arthurtok/decision-boundaries-visualised-via-python-plotly
        #get max and min for X and Y
        x_min, x_max = Xi[:, 0].min() - 1, Xi[:, 0].max() + 1
        y_min, y_max = Xi[:, 1].min() - 1, Xi[:, 1].max() + 1

        #Plot these x and y min max on grid
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                                 np.arange(y_min, y_max, h))
        
        Z = knni.predict(np.c_[xx.ravel(), yy.ravel()])
        # Put the result into a color plot
        Z = Z.reshape(xx.shape)    
        p = ax.pcolormesh(xx, yy, Z, cmap=ListedColormap(('orange', 'cornflowerblue')), alpha=0.35)

        # Plot also the training points
        ax.scatter(Xi[:, 0], Xi[:, 1], label=str(i), c=Yi, s=20)
        ax.set(xlim=[xx.min(), xx.max()], ylim=[yy.min(), yy.max()])
        plt_idx += 1 #incremeent the plot number for next iteration

fig.text(0.2, 0.0, 'k=1', ha='left')
fig.text(0.5, 0.0, 'k=25', ha='center')
fig.text(0.8, 0.0, 'k=50', ha='right')
fig.text(0.0, 0.8, 'dataset=1', va='center', rotation='vertical')
fig.text(0.0, 0.5, 'dataset=25', va='center', rotation='vertical')
fig.text(0.0, 0.2, 'dataset=50', va='center', rotation='vertical')

fig.suptitle('Compairing KNN Classifier for K = (1,25,50)')
plt.show()
```

Across rows (different dataset), the decision boundary varies for different randomized points. For K=50, the decision boundary is highly inflexible and therefore very similar for all the dataset. For K=1, the decision boundary varies the most across different datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

train_set = np.genfromtxt('data/A2_Q7_train.csv',delimiter=",", skip_header=1)
test_set = np.genfromtxt('data/A2_Q7_test.csv',delimiter=",", skip_header=1)

X_train = train_set[:,1]
y_train = train_set[:,2]

X_test = test_set[:,1]
y_test = test_set[:,2]

fig = plt.figure(figsize=(10, 10))
fig.suptitle('Scatterplot of Training Data')
fig.text(0.5, 0.05, 'X', ha='right')
fig.text(0.05, 0.5, 'Y', va='center', rotation='vertical')
plt.scatter(X_train,y_train)
plt.show()
```

```python
#b
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)

X_test = X_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)

lm1 = linear_model.LinearRegression()
lm1.fit(X_train, y_train)

# predictions
y_pred = lm1.predict(X_train)

print('R-Squared: {}'.format(r2_score(y_train, y_pred)))
print("Mean squared error: {}".format(mean_squared_error(y_train, y_pred)))

print('Coefficients: {}'.format(lm1.coef_))
print('Intercept: {}'.format(lm1.intercept_))
```

```python
#c
# Squared term -> y_hat = b_0 + b_1*x + b_2*x^2
X_train_sq = np.column_stack((X_train, np.power(X_train, 2)))
X_test_sq = np.column_stack((X_test, np.power(X_test, 2)))

lm2 = linear_model.LinearRegression()
lm2.fit(X_train_sq, y_train)

# predictions
y_pred_sq = lm2.predict(X_train_sq)

print('R-Squared: {}'.format(r2_score(y_train, y_pred_sq)))
print("Mean squared error: {}".format(mean_squared_error(y_train, y_pred_sq)))

print('Coefficients: {}'.format(lm2.coef_))
print('Intercept: {}'.format(lm2.intercept_))
```

```python
#c
# Squared term -> y_hat = b_0 + b_1*x + b_2*x^2
X_train_sq_cb = np.column_stack((X_train, np.power(X_train, 2), np.power(X_train, 3)))
X_test_sq_cb = np.column_stack((X_test, np.power(X_test, 2), np.power(X_test, 3)))

lm3 = linear_model.LinearRegression()
lm3.fit(X_train_sq_cb, y_train)

# predictions
y_pred_sq_cb = lm3.predict(X_train_sq_cb)

print('R-Squared: {}'.format(r2_score(y_train, y_pred_sq_cb)))
print("Mean squared error: {}".format(mean_squared_error(y_train, y_pred_sq_cb)))

print('Coefficients: {}'.format(lm3.coef_))
print('Intercept: {}'.format(lm3.intercept_))
```

```python
#d
fig = plt.figure(figsize=(10, 10))
plt.scatter(X_train,y_train)

x = np.linspace(min(X_train), max(X_train), 1000)

plt.plot(x, lm1.coef_[0][0]*x + lm1.intercept_[0], label='Linear model', color='g')
plt.plot(x, lm2.coef_[0][0]*x + lm2.coef_[0][1] *x**2 + lm2.intercept_[0], label='Tranformed model 1', color='b')
plt.plot(x, lm3.coef_[0][0]*x + lm3.coef_[0][2]*x**3 + lm3.coef_[0][1] *x**2 + lm3.intercept_[0], label='Transformed model 2', color='r')

fig.suptitle('Scatterplot of Training Data')
fig.text(0.5, 0.05, 'X', ha='right')
fig.text(0.05, 0.5, 'Y', va='center', rotation='vertical')
plt.legend()

plt.show()
```

```python
# predictions
y_pred_test = lm2.predict(X_test_sq)

print('R-Squared: {}'.format(r2_score(y_test, y_pred_test)))
print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred_test)))
```

### (f)

The tranformed model performs better on both training and testing data. The linear model does not fit the data at all and the R square value is negative for both training and testing. For second model with only one polynomial term, the model is slightly better for training data (R square becomes positive) but for testing data it performs poorly. This is an indication that the model is overly generlaized and therefore underfitting the data. 

The third model with two polynomial terms, performs relatively better on both training and testing data with significant improvement in R square. However, if we compare among performance of this model on training and testing data, the model provides better perforance from the testing data. This indicates that the model with higher R square generalize well with respect to x. 

### (g)

If the testing data is significantly different from the training data, the model might not generalize well and maybe result in overfitting. The predictive capability of the model will be significantly reduced in such situation. In such scenarios, as stated in (f), the model is trying to hard to find patterns in training data set and might indicate better performance on only traning data set. 

The idea is to have a model which generalizes well on testing (unseen) data and does not try very hard to find patterns on only training data set.
